Applying Data Cleaning on the dataset


```python
for x in df.index:
    if df.loc[x, "Status"] ==  "Removed":
        df.drop(x, inplace=True)
    
```

```python
df.shape
#WE dropped the Removed Roller Coasters data and hence shape is changed as rows
# have certainnly decreased.
```

```python
# Lets change the operating date

df["Opening date"] = pd.to_datetime(df["Opening date"],format='mixed', errors='coerce')

```

```python
df["Date","Month","Year"] = df["Opening date"]
#This is not a respetful way to represent a data
```

```python
df
```

```python
df["Date"] = df["Opening date"].dt.day
df["Month"] = df["Opening date"].dt.month
df["Year"] = df["Opening date"].dt.year
```

```python
df.drop(columns=[('Date', 'Month', 'Year')],inplace =True)
```

```python
print(df.columns)
```

```python
df.columns
```

```python
# df["Height restriction"]=df["Height restriction"].str.replace("NaN","Not Available")
```

```python
df["Location"]=df["Location"].fillna("-")
```

```python
df["Location"] = df["Location"].str.replace("Other", "Not Specified")
```

```python
df[['City', 'Country']] = df['Location'].str.split(', ', n=1, expand=True)
```

```python
df
```

```python
df = df.drop(columns = "Location")
```

Here we will remove the country and city from the end and place them in the start 
where location was placed

```python
country_column = df.pop("Country")
city_column = df.pop("City")
```

```python
df.insert(1,"Country",country_column)
df.insert(2,"City",city_column)
```

```python
manufacturer_counts = df['Manufacturer'].value_counts()
manufacturer_counts = manufacturer_counts.head(10)

# Plotting
plt.figure(figsize=(10, 4))  # Adjust figure size as needed
plt.bar(manufacturer_counts.index, manufacturer_counts.values)
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.title('Number of Roller Coasters by Manufacturer')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
# plt.tight_layout()
plt.show()
```

```python
Country_Count = df["Country"].value_counts()
Country_Count =Country_Count.head(10)
plt.figure(figsize=(10,6))
plt.xlabel = ("Countries")
plt.ylabel = ("Values")
plt.title("Most Roller Coasters by countries")
plt.bar(Country_Count.index, Country_Count.values)
plt.xticks(rotation=40)  # Rotate x-axis labels for better readability if needed
plt.tight_layout()
plt.show()
plt.show()
```

```python
df= df.dropna(subset=["speed_mph","height_value"])
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

# Plot "speed_mph"
plt.plot(df.index, df['speed_mph'], marker='o', linestyle='-', color='b', label='Speed (mph)')

# Plot "height_value"
plt.plot(df.index, df['height_value'], marker='o', linestyle='-', color='g', label='Height')

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Behavior of Speed and Height over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#From this figure we will se that there exsists a outlier having a very High value
```
